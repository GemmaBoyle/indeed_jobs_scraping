```python
import requests
import bs4
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from tqdm import tqdm
from random import randint
from time import sleep
```

```python
df = pd.read_csv('AU_jobs.csv')
```

```python
df = df.drop(columns='Unnamed: 0')
```

```python
df = df.drop_duplicates().dropna().reset_index(drop=True)
```

```python
df
```

```python
df = df[~df.salary.str.contains("week","month")]
```

```python
df = df[~df.salary.str.contains("hour")]
```

```python
yearly_salary = df.reset_index(drop=True)
```

```python
def strip_salary(x):
    nums = []
    y = x.replace('$','').replace(',','')
    for i in y.split():
        if i.isnumeric():
            nums.append(int(i))
    if len(nums) == 2:
        return 0.5*(nums[0]+nums[1])
    else:
        return nums[0]
```

```python
yearly_salary['salary'] = yearly_salary.salary.map(strip_salary)
```

```python
def convert_aud_to_gbp(amount):
    return amount*0.55
```

```python
yearly_salary['salary'] = yearly_salary.salary.map(convert_aud_to_gbp)
```

```python
au_df = yearly_salary
```

```python
au_df['country_location'] = 'AU'
```
