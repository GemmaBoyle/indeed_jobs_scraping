```python
import pandas as pd
```

```python
df = pd.read_csv('UK_jobs.csv')
df.count()
```

```python
df2 = pd.read_csv('UK_remote.csv')
df2.count()
```

```python
frames = [df, df2]

result = pd.concat(frames)

result
```

```python
df = result.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
df.count()
df
```

```python
df = df.drop(columns='Unnamed: 0')
```

```python
df = df.drop_duplicates().dropna()
```

```python
df.info()
```

```python
df.head(10)
```

```python
year_salary = df[~df.salary.str.contains("week","month")]
```

```python
year_salary
```

```python
def avg_salary(salary_snippet):
    x = salary_snippet.split()
    try:
        a = x[0].strip('£').replace(',', '').replace('.','')
        b = x[2].strip('£').replace(',', '').replace('.','')
        return 0.5*(int(a)+int(b))
    except: 
        return int(x[0].strip('£').replace(',', '').replace('.',''))
```

```python
year_salary['salary'] = year_salary.salary.map(avg_salary)
```

```python
year_salary
```

```python
year_salary['country_location'] = 'UK'
```

```python
year_salary
```

```python
year_salary.to_csv('uk_df')
```
