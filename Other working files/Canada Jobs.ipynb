```python
import requests
import bs4
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from tqdm import tqdm
from random import randint
from time import sleep
```

```python
def find_job_title(result):
    try: 
        return result.find(class_='jobTitle').text
    except:
        return np.nan
```

```python
def find_company_name(result):
    try:
        return result.find(class_='companyName').text
    except: 
        return np.nan
```

```python
def find_company_location(result):
    try:
        return result.find(class_='companyLocation').text
    except: 
        return np.nan
```

```python
def find_job_snippet(result):
    try:
        return result.find(class_='job-snippet').text.strip('\n')
    except:
        return np.nan
```

```python
def find_salary(result):
    try:
        return result.find(class_='salary-snippet').text
    except:
        return np.nan
```

```python
job_title = []
company_name = []
company_location = []
salary = []
snippet = []
```

```python
for start in tqdm(range(0, 1764, 16)):
        url = "https://ca.indeed.com/jobs?q=data+science&start={}".format(start)
        r = requests.get(url)

        sleep(randint(10,20))
        
        soup = BeautifulSoup(r.text, 'html.parser')
        for i in soup .find_all('div', class_='job_seen_beacon'):
            job_title.append(find_job_title(i))
            company_name.append(find_company_name(i))
            company_location.append(find_company_location(i))
            salary.append(find_salary(i))
            snippet.append(find_job_snippet(i))
```

```python
CA_jobs = pd.DataFrame(list(zip(job_title, company_name, company_location, salary, snippet)),
               columns =['job_title', 'company_name', 'company_location', 'salary','job_snippet'])

CA_jobs.head()
CA_jobs.count()
```

```python
CA_jobs_no_dup = CA_jobs.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
CA_jobs_no_dup.count()
```

```python
CA_jobs_no_dup.to_csv('CA_jobs.csv')
```

```python
df = pd.read_csv('CA_jobs.csv')
```

```python
df = df.drop(columns='Unnamed: 0')
```

```python
df = df.drop_duplicates().dropna().reset_index(drop=True)
df.head(35)
```

```python
yearly_salary = df.dropna()
```

```python
yearly_salary.info()
```

```python
yearly_salary.reset_index(drop=True)
```

```python
yearly_salary = yearly_salary[~yearly_salary.salary.str.contains("week","month")]
```

```python
yearly_salary = yearly_salary[~yearly_salary.salary.str.contains("hour")]
```

```python
yearly_salary = yearly_salary.reset_index(drop=True)
```

```python
def strip_salary(x):
    nums = []
    y = x.replace('$','').replace(',','')
    for i in y.split():
        if i.isnumeric():
            nums.append(int(i))
    if len(nums) == 2:
        return 0.5*(nums[0]+nums[1])
    else:
        return nums[0]
```

```python
yearly_salary['salary'] = yearly_salary.salary.map(strip_salary)
```

```python
yearly_salary.head()
```

```python
def convert_aud_to_gbp(amount):
    return amount*0.59
```

```python
yearly_salary['salary'] = yearly_salary.salary.map(convert_aud_to_gbp)
```

```python
ca_df = yearly_salary
```

```python
ca_df['country_location'] = 'CA'
```

```python
ca_df.to_csv('ca_df')
```
