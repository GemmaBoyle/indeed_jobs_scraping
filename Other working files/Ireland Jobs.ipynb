#### Find the entries with annual salary entries, by filtering the entries without salaries or salaries that are not yearly (filter those that refer to hour or week). Also, remove duplicate entries.

```python
df = df.dropna()
df
```

```python
year_salary = df[~df.salary.str.contains("week","month")]
```

```python
year_salary
```

```python
x = df.salary[0]
x
```

```python
def avg_salary(salary_snippet):
    x = salary_snippet.split()
    try:
        a = x[0].strip('€').replace(',', '')
        b = x[2].strip('€').replace(',', '')
        return 0.5*(int(a)+int(b))
    except: 
        return int(x[0].strip('€').replace(',', ''))
```

```python
avg_salary(x)
```

```python
year_salary['salary'] = year_salary.salary.map(avg_salary)
```

```python
year_salary
```

```python
def convert_eu_to_gbp(amount):
    return amount*0.85
```

```python
year_salary['salary'] = year_salary.salary.map(convert_eu_to_gbp)
```

```python
ie_df = year_salary
```

```python
ie_df['country_location'] = 'IE'
```

```python
# Convert the cleaned and converted data back to a df 

ie_df.to_csv('ie_df')
```
