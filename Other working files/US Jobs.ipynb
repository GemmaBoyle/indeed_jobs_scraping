```python
import pandas as pd
```

```python
df1 = pd.read_csv('US_jobs.csv')
```

```python
df2 = pd.read_csv('US_jobs(San_Jose_Remote.csv')
```

```python
df3 = pd.read_csv('US_jobs(NY__to_Atlanta.csv')
```

```python
df4 = pd.read_csv('US_jobs(Houston_to_sandiego.csv')
```

```python
df5 = pd.read_csv('US_jobs(Dallas_to_Denver.csv')
```

```python
frames = [df1, df2, df3, df4, df5]

result = pd.concat(frames)

result.head(5)
```

```python
df = result.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
df.count()
```

```python
df = df.drop(columns='Unnamed: 0')
```

```python
df = df.drop_duplicates().dropna()
```

```python
df.info()
```

```python
year_salary = df[~df.salary.str.contains("week","month")]
```

```python
year_salary = year_salary.reset_index(drop=True)
```

```python
year_salary = year_salary[~year_salary.salary.str.contains('hour')]
```

```python
year_salary.reset_index(drop=True)
```

```python
def strip_salary(x):
    nums = []
    y = x.replace('$','').replace(',','')
    for i in y.split():
        if i.isnumeric():
            nums.append(int(i))
    if len(nums) == 2:
        return 0.5*(nums[0]+nums[1])
    else:
        return nums[0]
```

```python
year_salary['salary'] = year_salary.salary.map(strip_salary)
```

```python
year_salary
```

```python
def convert_usd_to_gbp(amount):
    return amount*0.74
```

```python
year_salary['salary'] = year_salary.salary.map(convert_usd_to_gbp)
```

```python
us_df = year_salary
```

```python
us_df['country_location'] = 'US'
```

```python
us_df.to_csv('us_df')
```
